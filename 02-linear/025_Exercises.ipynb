## 2. Implement based on the Ridge regression example, the Lasso regression.

Please implement the SGD method and compare the results with the sklearn Lasso regression results. 

```python
def sgd2(w1, w0, X, y, alpha):
    # N
    n_obs = len(X)

    # cost function derivative for w1 and w0
    deriv_1 = deriv_0 = 0
    for i in range(1, n_obs):  
        deriv_1 += -2*X[i,1]*(y[i] - (w1*X[i,1] + w0))
        deriv_0 += y[i] - (w1*X[i,1] + w0)

    w1_new = w1 - alpha*deriv_1
    w0_new = w0 - alpha*deriv_0 

    return w1_new, w0_new
```

```python
x = np.array([188, 181, 197, 168, 167, 187, 178, 194, 140, 176, 168, 192, 173, 142, 176]).reshape(-1, 1).reshape(15,1)
y = np.array([141, 106, 149, 59, 79, 136, 65, 136, 52, 87, 115, 140, 82, 69, 121]).reshape(-1, 1).reshape(15,1)

x = np.asmatrix(np.c_[np.ones((15,1)),x])

alpha = 0.000000001 

w1 = w0 = 1

for i in range(10000):
    w1, w0 = sgd2(w1, w0, x, y, alpha)
print(w1, w0)
```

## 3. Extend the Fisher's classifier

Please extend the targets of the ``iris_data`` variable and use it as the $y$.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris_data = load_iris()
iris_df = pd.DataFrame(iris_data.data,columns=iris_data.feature_names)
iris_df.head()

x = iris_df['sepal width (cm)'].values
y = iris_df['sepal length (cm)'].values 

dataset_size = np.size(x)

mean_x, mean_y = np.mean(x), np.mean(y)
```

```python
# variances
s_x = 1/(dataset_size-1) + np.sum((x-mean_x)*(x-mean_x).T)
s_y = 1/(dataset_size-1) + np.sum((y-mean_y)*(y-mean_y).T)

s_w = 1/(dataset_size+dataset_size-2)*((dataset_size-1)*s_x*(dataset_size-1)*s_y)

# weight
w = s_x**-1*(mean_x-mean_y)
```

```python
# g(x) = wT*x - w0
G_x = w.T*x - 1/2*(w.T*mean_x+w.T*mean_y)
```

```python
print(s_x, s_y, s_w, w)
```

```python
# use it as the y?
x2 = y
y = G_x

mean_x, mean_x2, mean_y = np.mean(x), np.mean(x2), np.mean(y)

SS_xy = np.sum(y * x) - dataset_size * mean_y * mean_x
SS_xx = np.sum(x * x) - dataset_size * mean_x * mean_x

SS_x2y = np.sum(y * x2) - dataset_size * mean_y * mean_x2
SS_x2x2 = np.sum(x2 * x2) - dataset_size * mean_x2 * mean_x2

a1 = SS_xy / SS_xx
a2 = SS_x2y / SS_x2x2

b = mean_y - a1 * mean_x - a2 * mean_x2

# y = w1*x1 + w2*x2 + w0
y_pred = a1 * x + a2 * x2 + b
```
